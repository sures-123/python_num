**Import Libraries**

```python
import pandas as pd
import numpy as np
```

**Read Excel**

```python
df=pd.read_excel("/content/report (4).xlsx")
df
```

**Remove Empty Rows & Columns**

```python
df.dropna(axis=1, how='all', inplace=True)
df.dropna(axis=0, how='all', inplace=True)
df
```

```python
df.shape
```

**Convert Excel To csv File**

```python
df.to_csv("file.csv")
```

**Assign Length**

```python
df.columns=np.arange(1,len(df.columns)+1)
```

**Delete top 1 row**

```python
df.drop(df.head(1).index,inplace=True)
df
```

```python
df.shape
```

**Remove Dollar Signs**

```python
df = df.replace('[$]', '', regex=True)
df
```

**Remove Braces**

```python
df = df.replace('[()]', '', regex=True)
df
```
